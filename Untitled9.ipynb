```python
import pandas as pd

targetUrl_iris="/content/sample_data/caffeine.csv"

# 아이리스 데이터 불러오기.
df = pd.read_csv(targetUrl_iris)
```

```python
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 그래프로 확인.
sns.pairplot(df.iloc[:,1:6], hue='type');
plt.show()
```

```python
# 속성을 X, 클래스를 y로 저장.
X = df.iloc[:,1:4]
y = df.iloc[:,4]

# X와 y의 첫 5줄을 출력.
print(X[0:5])
print(y[0:5])
```

```python
# 원-핫 인코딩 처리.
y = pd.get_dummies(y)

# 원-핫 인코딩 결과 확인.
print(y[0:5])
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

import numpy as np
X = np.asarray(X).astype(np.float32)

# 모델 설정
model = Sequential()
model.add(Dense(60, input_dim=3, activation='relu'))
model.add(Dense(30, activation='relu'))
model.add(Dense(6, activation='softmax'))

# 요약
model.summary()

# 모델 컴파일
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# 모델 실행
history=model.fit(X, y, epochs=30, batch_size=10)
```

```python
model = Sequential()
model.add(Dense(30, input_dim=3, activation='relu'))
model.add(Dense(10, activation='sigmoid'))
model.add(Dense(6, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['binary_accuracy'])
history=model.fit(X, y, epochs=100, batch_size=6)
```
