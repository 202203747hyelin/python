```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from sklearn.model_selection import train_test_split

import matplotlib.pyplot as plt
import seaborn as sns

import pandas as pd
import numpy as np

df = pd.read_csv("/content/sample_data/heart disease classification dataset.csv")

df
```

```python
df.dtypes
```

```python
df.isnull().sum().sort_values(ascending=False).head(20)
```

```python
df = pd.get_dummies(df)
 
df = df.fillna(df.mean())

df
```

```python
df_corr=df.corr()

df_corr_sort=df_corr.sort_values('thal', ascending=False)

df_corr_sort['thal'].head(10)
```

```python
cols=['thal','target_no','Unnamed: 0','oldpeak','sex_male','exang']
sns.pairplot(df[cols])
plt.show();
```

```python
cols_train=['target_no','Unnamed: 0','oldpeak','sex_male','exang']
X_train_pre = df[cols_train]

y = df['thal'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_train_pre, y, test_size=0.1)
```

```python
model = Sequential()
model.add(Dense(10, input_dim=X_train.shape[1], activation='relu'))
model.add(Dense(30, activation='relu'))
model.add(Dense(40, activation='relu'))
model.add(Dense(1))
model.summary()

model.compile(optimizer ='adam', loss = 'mean_squared_error')

early_stopping_callback = EarlyStopping(monitor='val_loss', patience=20)

modelpath="./data/model/Ch15-house.hdf5"

checkpointer = ModelCheckpoint(filepath=modelpath, monitor='val_loss', verbose=0, save_best_only=True)

history = model.fit(X_train, y_train, validation_split=0.25, epochs=2000, batch_size=32, callbacks=[early_stopping_callback, checkpointer])
```

```python
real_disease =[]
pred_disease = []
X_num = []

n_iter = 0
Y_prediction = model.predict(X_test).flatten()
for i in range(25):
    real = y_test[i]
    prediction = Y_prediction[i]
    print("실제발병: {:.2f}, 예상발병: {:.2f}".format(real, prediction))
    real_disease.append(real)
    pred_disease.append(prediction)
    n_iter = n_iter + 1
    X_num.append(n_iter)
```

```python
plt.plot(X_num, pred_disease, label='predicted disease')
plt.plot(X_num, real_disease, label='real disease')
plt.legend()
plt.show()
```
